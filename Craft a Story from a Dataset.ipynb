# Craft a Story from a Dataset

### Import data

```python
with open('temps.txt') as f:
    lines = f.readlines()
f.close()
```

```python
lines
```

```python
import pandas as pd

df= pd.DataFrame(lines)
df
```

```python
df=df[0].str.split('\t', expand=True)
```

```python
df.columns = df.iloc[0] 

df = df[1:]

df
```

## The question

### Using the data of daily high temperature data for Atlanta (July through October), has the unofficial end of summer been getting  later?

## Investigating trends

```python
import matplotlib.pyplot as plt

```

```python
df=df.drop(['2015\n'], axis=1)
```

```python
df.dtypes
```

```python
df.plot(kind='bar', x='DAY', figsize=(20,20))
```

```python
#My conclusion from the bar graph above is that July is hotter than October.
```

```python
df2 = df
df2=df2.drop(['1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013'], axis=1)
df2.plot(kind='bar', x='DAY', figsize=(20,20))
```

### Looking at the tail of the 1996 and 2014 data, a convincing argument can be made that the unofficial end of summer has gotten later over the years.
