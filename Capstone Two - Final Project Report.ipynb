# 5. Predictions and Accuracy

### 5.1 The Association Model

####  The antecedents and the consequents are the pairs. 
####  Support = freq(A,B) / N
###### Support measures the frequency of the items A and B. It shows the items bought more frequently. 
#### Confidence = freq(A,B) / freq(A)
###### Confidence measures how often the items A and B occur togther given the number of times A occur.
####  Lift  = Support / (Supp(A) * Supp(B))
###### Lift takes into account the randomness of the system. It is the support of A and B divided by the independent occurence probability of A and B. This gives the strength of Items A and B together while taking into account the occurence of randomness.

#####  The pairs with the best support are bread and coffee, cake and coffee, and tea and coffee. This is not surprising  as these are the most frequently bought items and the fact that coffee is in all of the pairs because it is by far the most frequently purchased item. 

#####  The pairs with the best confidence are toast and coffee, spanish brunch and coffee, and medialuna and coffee. I should mention that confidence does not take into cosideration the freq(B) in the denominator which would be coffee for the above pairs.

#####  Sticking with pairs of two, the pairs with the best lift are cake and tea, toast and coffee, and spanish brunch and coffee. I am curious as to why the lift for coffee, tea, and cake is higher than the lift of those items with pairs of two.
